# Introdução as f-strings(Formatação de strings):
- Formatação básica de strings.
- `s - string`
- `d - int`
- `f - float`
- `x ou X - hexadecimnal`
- `(caractere)(<>^)(quantidade)`
- `> - esquerda`
- `< - direita`
- `^ - centro`
- Sinal:  + ou -

> Ex: `.: 0>-100,.1f`

- Conversion flags: `!r`, `!s`, `!a`
- métodos __repr__, __str__ e __ask__

> Quando uma variável requer um tamanho fixo e ele não é satisfeito, criamos uma largura fixa para essa variável. Isso se da o nome de `Pad`.

```python
variavel = 'ABC'
print(f'{variavel}')
# ou seja, a variável terá 10 espaços, sendo 3 de 'ABC' e 7 do auto preenchimento.
print(f'{variavel: >10}')
print(f'{variavel: <10}')
print(f'{variavel: ^10}.')
print(f'{variavel:$^10}.')
print(f'{1000.45195419849:,.1f}')
print(f'{-1000.45195419849:+.1f}')
# quando for positivo eu quero que o valor de + apareça.
print(f'{1000.45195419849:+.1f}')
print(f'O hexadecimal de 1500 é {1500:08X}')
print(f'O hexadecimal de 1500 é {1500:04X}')
print(f'{variavel!r}')

# nome = 'Jorge Manoel'
# altura = float(1.80)
# peso = float(95)
# imc = peso / (altura * altura)
# # imc = ...

# linha_1 = (f'{nome} tem {altura} de altura. Seu IMC é de {imc:.2f}')
# print(linha_1)

# print(nome, 'tem', altura, 'de altura')
# print('Pesa', peso, 'quilos')
# print('O seu IMC é de', imc)
```

# Formatação com o método format:

- O `'a', 'b', 'c' tem índices`. Sendo assim, não necessariamente é obrigatório colocar na ordem ou apenas uma vez o valor para formatar. Podemos utilizar diretamente o número do índice dentro do local do `format`. Assim, podemos evitar o erro de `Out of Range`.
- Podemos também nomear o parâmetro, como no exemplo abaixo. Devemos ter atenção pois se nomearmos um parâmetro, todos os seguintes precisam também ser nomeados.
- a, b e c são argumentos. nome1 é um `parâmetro do argumento`.

```python
a = 'AAAAA'
b = 'BBB'
c = 1.1

# chamada pelo nome do parâmetro
string1 = 'a={} b={} c={nome3:.2f}'
formato1 = string1.format(a, b, nome3=c)

# chamada pelo índice do parâmetro.
string2 = 'c={2:.2f} b={1} b={1} a={0}'
formato2 = string2.format(a, b, c)

print(formato1)
print(formato2)
```

# Função `input` para coletar dados do usuário:

- Sempre o input receberá type strings.
- Ex: `numero2 = input('Digite outro número: ')`.
- Podemos também fazer a conversão de tipo direto no input, desta forma: `numero2 = int(input('Digite outro número: '))`. Porém, este formato pode ocasionar problemas futuros.
- Um exemplo é, se o usuário digitar 'A' em vez de um número vamos ter um erro logo de cara.
- O ideal é sempre utilizarmos a conversão de tipo criando outra variável para converter. Isto é importante para checar o que o usuário digitou.

```python
nome = input('Qual o seu nome: ')
numero1 = input('Digite um número: ')
int_numero1 = int(numero1)
numero2 = input('Digite outro número: ')
int_numero2 = int(numero2)

print(f'O seu nome é {nome =}') # o = dentro do argumento faz com que não precisemos adicionar o '=' fora dele.
print(f'A soma dos números é: {numero1 + numero2}') # neste caso ele vai concatenar pois se tratam de duas strings.
print(f'A soma dos números é: {int_numero1 + int_numero2}') # neste caso ele vai concatenar pois se tratam de duas strings.
```

# Operadores condicionais:
- `if: se - elif - se não, se - else: se não`
- Altera o fluxo do código.
- Podemos utilizar apenas o `if` se necessário, porém o `elif` e `else` não é possível.
- Assim que o bloco de código acha a alternativa verdadeira ele sai do bloco e vai para o próximo.

```python
entrada = input('Você quer "entrar" ou "sair"?\n')

if entrada == 'entrar':
  print('Você entrou no programa.')
elif entrada != 'entrar' or 'sair':
  print('Você digitou algo inválido.')
else:
  print('Você saiu do programa.')

if 10 == 10:
  print('Outro if')
```

- No exemplo abaixo temos a condição 1 e 2 como `False`. Então, o código vai pular esta parte. Porém, na condição 3 temos um `True`, sendo assim, o código vai parar nele. Independente da condição 4 ser True também.

```python
condicao1 = False
condicao2 = False
condicao3 = True
condicao4 = True

if condicao1:
  print('Código para condição 1')
elif condicao2:
  print('Código para condição 2')
elif condicao3:
  print('Código para condição 3')
elif condicao4:
  print('Código para condição 4')
else:
  print('Nenhuma condição foi satisfeita.')

if 10 == 10:
  print('Outro if')
```

# Operadores relacionais (de comparação):
> Operadores:
- Sempre que usarmos qualquer um dos operadores de comparação eles retornam um `boolean(True or False)`.
- `>, <, >=, <=, ==, !=`.
> Exemplo:

```python
primeiro_valor = input('Digite um valor numérico: ')
segundo_valor = input('Digite um valor numérico: ')

if primeiro_valor > segundo_valor:
  print(f'O maior número é o {primeiro_valor} e o menor é o {segundo_valor}')
elif segundo_valor > primeiro_valor:
  print(f'O maior número é o {segundo_valor} e o menor é o {primeiro_valor}')
else:
  print('Os números são iguais')
```

# Operadores Lógicos: `and(e)`, `or(ou)` e `not(não)`.
- `and`: Todas as condições precisam ser verdadeiras. Se qualquer valor for considerado falso, a expressão inteira será avaliada naquele valor. São considerados `falsy`. 0, 0.0 ou ' ' são `False`.

- `or`: Qualquer condição verdadeira avalia toda a expressão como `True`. Se qualquer valor for considerado verdadeiro, a expressão inteira será avaliada naquele valor.

# Operadores `in(está entre)` e `not in(não está entre)`:

- Strings em python são iteráveis. Ou seja, podemos navegar item por item.
- `0 1 2 3 4 5 6` (índices positivos)

- `W-I-L-L-I-A-M`

- `-7-6-5-4-3-2-1` (índices negativos)

```python
nome1 = 'William'

print(nome1[2])
print('Will' in nome1)
print('l' in nome1)
print('0' in nome1)
print('x' not in nome1)
print('m' not in nome1)

print(10 * '-')

nome2 = input('Digite o seu nome: ')
# abreviação para não utilizar um if
encontrar = input('Digite o que deseja encontrar: ') in nome2
print(encontrar)

print(10 * '-')

nome3 = input('Digite o seu nome: ')
encontrar = input('Digite o que deseja encontrar: ')
if encontrar in nome3:
  print(f'Em {nome3} existe {encontrar}')
else:
  print(f'Em {nome3} não existe {encontrar}')

print(10 * '-')

nome4 = 'Maria Carmo'
if ' ' in nome4:
    print(f'O nome {nome4} tem espaços.')
else:
    print(f'O nome {nome4} NÃO tem espaços.')
```

# Interpolação básica de strings:

- interpolação, por exemplo, é o que fazemos com o format.
- Precisamos criar a string, colocar antes da string o '`%`' e passar os valores para preencher (`place holders`).

- `s - string`
- `d e i - int`
- `f - float`
- `x e X - Hexadecimal`(0123456789ABCDEF)

- Em resumo, podemos escolher qual o tipo queremos utilizar. Ou utilizarmos a `f-string` ou o `format` ou a `interpolação`.

```python
nome = 'Luiz'
preco = 1000.95897643

# podemos fazer o que está abaixo de outra forma
# variavel = 'Luiz, o preço total foi R$1000.95'

variavel1 = '%s, o preço total foi R$' % nome
print(variavel1)

# parâmetro para casas decimais
variavel2 = '%s, o preço total foi R$%.2f' % (nome, preco)
print(variavel2)

print(10 * '-')

print('O hexadecimal de %d é %x' % (15, 15))
print('O hexadecimal de %d é %X' % (15, 15))
print('O hexadecimal de %d é %04x' % (1500, 1500))
print('O hexadecimal de %d é %04X' % (1800, 1800))
```

# Fatiamento de strings:

- Podemos pegar o início, o fim e o passo.
- `[i:f:p]` e `[::](string inteira)`
- A função `len` retorna a quantidade de caracteres da str.

```python
variavel = 'abcdefghijklmnopqrstuvxz'
print(variavel[4:])
print(variavel[:4])
print(variavel[:8:2])
print(variavel[::-1])
print(variavel[-1:-25:-1])
print(variavel[-1:-25:-2])
print(len(variavel))
```
