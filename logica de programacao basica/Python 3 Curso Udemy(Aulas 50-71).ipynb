# Variáveis, Constantes e Complexidade do código.

- Os códigos são escritos para nós e para outras pessoas, não para a máquina. Então, precisamos fazer de uma forma simples e coompreensível, Para que outras pessoas ou até nós mesmos depois consigamos entender o que está digitado.

- Táticas para escrever um código mais limpo:
  
  - `Constantes`: Variável que não muda. Geralmente utilizamos para constantes `LETRAS_MAIUSCULAS` para definir. Isso é uma convenção de escrita de código.
  
  - Quanto mais condições tivermos dentro de um if mais complicado vai ser para entender. Devemos evitar isso.

  - Quanto mais afastado da margem o seu código está, ou seja, quanto mais códigos dentro de códigos existir, mais complexo vai ser. E isso não é uma boa prática para criar códigos limpos.

````python
velocidade_carro = 50 # velocidade atual do carro
local_carro = 99 # local em que o carro está na estrada

LIMITE_VEL_RADAR_1 = 60 # velocidade máxima do radar 1
LOCAL_RADAR_1 = 100 # local onde o radar 1 está
RANGE_RADAR_1 = 1 # a distância onde o radar pega

if velocidade_carro > LIMITE_VEL_RADAR_1:
  print('Velocidade do carro passou do radar 1.')

# Isso quer dizer que, se o carro passar acima de 60km/h entre o km 99 e 101, que é o range do radar, ele será multado.
if local_carro >= (LOCAL_RADAR_1 - RANGE_RADAR_1) and local_carro <= (LOCAL_RADAR_1 + RANGE_RADAR_1) and velocidade_carro > LIMITE_VEL_RADAR_1:
  print('O carro foi multado no radar 1.')
````

> Limpando o código:
- Para condições muito longas podemos criar variáveis para armazenar esse valor e jogar apenas a variável na condição.
- Variáveis não são para resumir códigos e sim para deixar mais legível.

````python
# Variáveis
velocidade_carro = 61
local_carro = 101

# Constantes
LIMITE_VEL_RADAR_1 = 60
LOCAL_RADAR_1 = 100
RANGE_RADAR_1 = 1

# Vai devolver o valor booleano se o carro estiver dentro do range do radar 1.
carro_passou_radar_1 = local_carro >= (LOCAL_RADAR_1 - RANGE_RADAR_1) and local_carro <= (LOCAL_RADAR_1 + RANGE_RADAR_1)

# Vai devolver o valor booleano se o carro exceder a velocidade permitida no radar 1.
vel_carro_passou_limite_radar_1 = (velocidade_carro > LIMITE_VEL_RADAR_1) and carro_passou_radar_1

# Vai devolver o valor booleano se o carro passar pelo range do radar e estiver acima da velocidade permitida dele.
carro_multado_radar_1 = carro_passou_radar_1 and vel_carro_passou_limite_radar_1

if carro_passou_radar_1:
  print('Carro passou radar 1')

if vel_carro_passou_limite_radar_1:
  print('Velocidade do carro excedeu o limite do radar 1.')

if carro_multado_radar_1:
  print('O carro foi multado no radar 1.')
````

# Id - A identidade do valor que está na memória

- `Flag(Bandeira)`: Marcar um local.

- `None`: Não valor.

- `is e is not`: É ou não é (tipo, valor, identidade).

- `id`: Identidade.

````python
v1 = 'a'
v2 = 'a'
v3 = 'b'
print(id(v1))
print(id(v2))
print(id(v3))

# criamos esse código para saber se o algoritmo em algum momento passou por aquela parte do código.
condicao = True
passou_no_if = None

if condicao:
  # Flag (Bandeira).
  passou_no_if = True
  print('Faça algo.')
else:
  print('Não faça algo.')

if passou_no_if is None:
  print('Não passou no if')

if passou_no_if is not None:
  print('Passou no if')

# passou_no_if is None é a mesma coisa de passou_no_if == None, Só que com o None utilizamos o is.
print(passou_no_if, passou_no_if is None)
print(passou_no_if, passou_no_if is not None)
````

# While(enquanto) e break - Estruturas de repetição.

- Executa uma ação enquanto a condição for verdadeira.
- Devemos ter cuidado com loops infinitos, por exemplo:
```
condicao = True

while condicao:
  nome = input('Digite o seu nome: ')
  print(f'Seu nome é {nome}')
```
- Para evitar isso devemos em algum momento mudar a condicao para `False` ou adicionar um `break`.

> Operadores de atribuição: `=`, `+=`, `-=`, `*=`, `/=`, `//=`, `**=`, `%=`.

````python
condicao = True

while condicao:
  nome = input('Digite seu nome: ')
  print(f'Seu nome é {nome}')

  if nome == 'sair':
    break

print('Acabou o loop.')
````

- É preciso ter atenção se o contador será adicionado antes ou depois da ação, pois ela pode mudar o resultado.

````python
indice = 0

while indice < 5:
  indice += 1
  print(indice)

print('---' *10)

indice = 0

while indice < 5:
  print(indice)
  indice += 1
````

- O `continue` serve para pular alguma parte do loop, por exemplo queremos pular alguns números e fazer ele voltar para o início, utilizamos o continue.

````python
indice = 0

while indice <= 10:
  indice += 1

  if indice == 4:
    print('Não vou mostrar o', indice)
    continue

  if indice >=7 and indice <=9:
    print('Não vou mostrar o', indice)
    continue

  print(indice)

print('Acabou o loop.')
````

# While + While (laços internos).

- Quando temos um while dentro de outro é como se fosse uma engrenagem. O primeiro while é a roda de fora, ele vai rodar até que chegue na condição da roda mais interna dele. A roda mais interna vai executar e finalizando a execução vai voltar para a roda maior, até que termine todas as execuções.

````python
qntd_linhas = 5
qntd_colunas = 5

linha = 1
while linha <= qntd_linhas:
  coluna = 1
  while coluna <= qntd_colunas:
    print(f'{linha=} {coluna=}')
    coluna += 1
  linha += 1

print('Acabou o loop.')
````

- Exercício de while:

````python
nome = 'Victoria Ferreira'
tamanho_nome = len(nome)
indice = 0
novo_nome = ''

while indice < tamanho_nome:
  novo_nome += nome[indice] + '*'
  indice += 1
print(novo_nome)

print('---' *10)

while indice < tamanho_nome:
  letra = nome[indice]
  novo_nome += f'{letra}*'
  indice += 1
print(novo_nome)
````

# Exercício de calculadora com while:
- Primeiro será solicitado o primeiro número ao usuário
- Depois pediremos o segundo número
- Depois perguntaremos a operação
- Só serão aceitas: adições, subtrações, multiplicações e divisões

> OBS: Da mesma forma que aplicamos a função `lower()` para que transforme o que for inputado em minúsculo, podemos utilizar outras funções também. Duas dessas funções importantes são `startswith(começa com)` e `endswith(termina com)`.

````python
sair = False

while not sair:

  primeiro_numero = input('Digite o primeiro número: ')
  segundo_numero = input('Digite o segundo número: ')
  print('---' * 10)

  try:
    primeiro_numero_int = int(primeiro_numero)
    segundo_numero_int = int(segundo_numero)

    operador = input("Informe o operador ou digite 'sair' para finalizar a operação.\nOperadores disponíveis: '+', '-', '*', '/'\nDigite a operação desejada: ").lower()

    if operador == '+':
      soma = primeiro_numero_int + segundo_numero_int
      print(f'\nResultado: {primeiro_numero_int} + {segundo_numero_int} = {soma}\n')

    elif operador == '-':
      subtracao = primeiro_numero_int - segundo_numero_int
      print(f'\nResultado: {primeiro_numero_int} - {segundo_numero_int} = {subtracao}\n')

    elif operador == '*':
      multiplicacao = primeiro_numero_int * segundo_numero_int
      print(f'\nResultado: {primeiro_numero_int} * {segundo_numero_int} = {multiplicacao}\n')

    elif operador == '/':
      divisao = primeiro_numero_int / segundo_numero_int
      print(f'\nResultado: {primeiro_numero_int} / {segundo_numero_int} = {divisao}\n')

    elif operador == 'sair':
      sair = True
      print('---' * 10)
      print('\nVocê optou por sair. Finalizando operação...')

    else:
      print('Operador desconhecido, tente novamente...')
      print('---' * 10)

  except:
    print('Você digitou um valor não reconhecível em algum dos campos acima. Tente novamente...')
    print('---' * 10)
````

# Código da calculadora do professor:

````python
while True:
    numero_1 = input('Digite um número: ')
    numero_2 = input('Digite outro número: ')

    numeros_validos = None

    try:
        num_1_float = float(numero_1)
        num_2_float = float(numero_2)
        numeros_validos = True
    except:
        numeros_validos = None

    if numeros_validos is None:
        print('Um ou ambos os números digitados são inválidos.')
        continue

    operadores_permitidos = '+-/*'

    operador = input('Digite o operador (+ - / *): ')

    if operador not in operadores_permitidos:
        print('Operador inválido.')
        continue

    if len(operador) > 1:
        print('Digite apenas um operador.')
        continue

    print('Realizando sua conta. Confira o resultado abaixo:')

    if operador == '+':
      print(f'{num_1_float} + {num_2_float} =', num_1_float + num_2_float)
    elif operador == '-':
      print(f'{num_1_float} - {num_2_float} =', num_1_float - num_2_float)
    elif operador == '*':
      print(f'{num_1_float} * {num_2_float} =', num_1_float * num_2_float)
    elif operador == '/':
      print(f'{num_1_float} / {num_2_float} =', num_1_float / num_2_float)
    else:
      print('Nunca deveria chegar aqui.')

    sair = input('Quer sair? [s]im: ').lower().startswith('s')

    if sair is True:
        break
````

# Exercício 1: Jogo "Adivinhe o Número"

> Crie um programa que escolhe um número secreto (por exemplo, 50). Em seguida, peça para o usuário adivinhar o número. O loop deve continuar até que o usuário acerte.

````python
numero_secreto = 50

while True:
  numero = input('Digite o número que acha que é: ')

  numero_valido = None

  try:
    numero_int = int(numero)
    numero_valido = True

  except:
    print('Você digitou um valor inválido. Tente novamente...')
    numero_valido = None
    continue

  if numero_int is not numero_secreto and numero_int < numero_secreto:
    print('Você está quase lá!\nO número que digitou é menor que o número secreto. Tente novamente...')
    continue

  if numero_int is not numero_secreto and numero_int > numero_secreto:
    print('Você está quase lá!\nO número que digitou é maior que o número secreto. Tente novamente...')
    continue

  else:
    print('Você acertou o número secreto. Parabéns!')
    break
````

# While - Qual letra apareceu mais vezes na frase:

- `count()`: função que vai retornar a quantidade de vezes que certa coisa apareceu em uma string.
- Para não termos problemas de minúsculos e maiúsculos podemos converter tudo para `lower()` ou `upper()`.


````python
frase = 'O Python é uma linguagem de programação '\
'multiparadigmaaa. '\
'Python foi criado por Guido van Rossum.'.upper()

i = 0
qtd_apareceu_mais_vezes = 0
letra_apareceu_mais_vezes = ''

frase_sem_espacos = frase.replace(' ', '')

while i < len(frase_sem_espacos):

  letra_atual = frase_sem_espacos[i]

  quantas_vezes_letra_atual_apareceu = frase_sem_espacos.count(letra_atual)

  # para pegar a última letra, caso ocorra um empate, podemos usar o >=
  if quantas_vezes_letra_atual_apareceu > qtd_apareceu_mais_vezes:
    qtd_apareceu_mais_vezes = quantas_vezes_letra_atual_apareceu
    letra_apareceu_mais_vezes = letra_atual

  i += 1

print(f'A letra que apareceu mais vezes no texto foi a letra "{letra_apareceu_mais_vezes}".\nEla apareceu {qtd_apareceu_mais_vezes} vezes no texto.')
````
